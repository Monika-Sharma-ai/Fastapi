## 1️⃣ Section: Import Required Libraries

First, let's import the necessary libraries for building FastAPI applications and testing them.

````python
# Import required libraries
from fastapi import FastAPI
from pydantic import BaseModel
from typing import Optional, List
import json

print("✅ All imports successful!")
print("- FastAPI: Web framework for building APIs")
print("- Pydantic: Data validation using Python type hints")
print("- Typing: Type hints for Python")
print("\nReady to build amazing APIs! 🎉")
````

## 2️⃣ Section: Create a Basic FastAPI Application

Let's initialize a FastAPI application instance. This is the core object that will manage all your routes and endpoints.

````python
# Create a FastAPI application instance
app = FastAPI(
    title="My First API",
    description="A simple FastAPI application for learning",
    version="1.0.0"
)

print("✅ FastAPI application created!")
print(f"Title: {app.title}")
print(f"Description: {app.description}")
print(f"Version: {app.version}")
print("\nYour FastAPI app is ready to receive routes!")
````

## 3️⃣ Section: Define API Routes with GET Requests

Now let's add our first routes! Routes are defined using decorators like `@app.get()`. Each route handles HTTP requests and returns responses.

````python
# Define a simple GET route at the root path
@app.get("/")
def read_root():
    """
    This is the root endpoint.
    It returns a simple greeting message.
    """
    return {
        "message": "Welcome to My First API!",
        "endpoint": "/",
        "method": "GET"
    }


# Define another GET route
@app.get("/api/status")
def get_status():
    """
    Get the current status of the API.
    """
    return {
        "status": "running",
        "api_version": "1.0.0",
        "message": "API is healthy and ready to serve requests"
    }


# Define a GET route that returns a list
@app.get("/api/items")
def get_items():
    """
    Get a list of items.
    """
    items = [
        {"id": 1, "name": "Item 1", "price": 9.99},
        {"id": 2, "name": "Item 2", "price": 19.99},
        {"id": 3, "name": "Item 3", "price": 29.99}
    ]
    return {"items": items, "count": len(items)}


print("✅ GET routes defined!")
print("Routes created:")
print("  - GET /")
print("  - GET /api/status")
print("  - GET /api/items")
````

## 4️⃣ Section: Define API Routes with POST Requests

POST requests are used to send data to the server. Let's define a Pydantic model first, which validates incoming data.

````python
# Define a Pydantic model for Item
class Item(BaseModel):
    """
    Pydantic model for an Item.
    This validates the structure of incoming JSON data.
    """
    name: str  # Required field
    price: float  # Required field
    description: Optional[str] = None  # Optional field with default value
    is_available: bool = True  # Optional field with default True


# Define a Pydantic model for User
class User(BaseModel):
    """Pydantic model for User"""
    id: Optional[int] = None
    name: str
    email: str
    age: int
    role: str = "user"  # Default role


# Define a POST route to create an item
@app.post("/api/items")
def create_item(item: Item):
    """
    Create a new item.
    The request body should contain: name, price, description (optional), is_available (optional)
    """
    return {
        "message": "Item created successfully",
        "item": item,
        "status": "created"
    }


# Define a POST route to create a user
@app.post("/api/users")
def create_user(user: User):
    """
    Create a new user.
    The request body should contain: name, email, age, role (optional)
    """
    return {
        "message": "User created successfully",
        "user": user,
        "timestamp": "2024-06-23T10:30:00Z"
    }


print("✅ POST routes defined!")
print("Routes created:")
print("  - POST /api/items")
print("  - POST /api/users")
print("\nPydantic Models:")
print("  - Item: name (str), price (float), description (optional), is_available (optional)")
print("  - User: name (str), email (str), age (int), role (optional)")
````

## 5️⃣ Section: Add Path Parameters

Path parameters are part of the URL. They allow you to pass dynamic values to your endpoints. For example, `/users/{user_id}` where `user_id` is a path parameter.

````python
# Mock database
items_db = {
    1: {"id": 1, "name": "Laptop", "price": 999.99},
    2: {"id": 2, "name": "Mouse", "price": 29.99},
    3: {"id": 3, "name": "Keyboard", "price": 79.99}
}


# GET route with a single path parameter
@app.get("/api/items/{item_id}")
def get_item_by_id(item_id: int):
    """
    Get a specific item by ID.
    Path parameter: item_id (integer)
    Example: GET /api/items/1
    """
    if item_id in items_db:
        return {
            "item": items_db[item_id],
            "found": True
        }
    else:
        return {
            "error": f"Item with ID {item_id} not found",
            "found": False,
            "status_code": 404
        }


# GET route with multiple path parameters
@app.get("/api/users/{user_id}/posts/{post_id}")
def get_user_post(user_id: int, post_id: int):
    """
    Get a specific post from a specific user.
    Path parameters: user_id (int), post_id (int)
    Example: GET /api/users/42/posts/7
    """
    return {
        "user_id": user_id,
        "post_id": post_id,
        "content": f"This is post {post_id} from user {user_id}",
        "title": f"Post {post_id}"
    }


# PUT route with path parameter to update an item
@app.put("/api/items/{item_id}")
def update_item(item_id: int, item: Item):
    """
    Update an item by ID.
    Path parameter: item_id (int)
    Body: Item model
    """
    if item_id in items_db:
        items_db[item_id] = {
            "id": item_id,
            "name": item.name,
            "price": item.price
        }
        return {
            "message": f"Item {item_id} updated successfully",
            "item": items_db[item_id]
        }
    else:
        return {
            "error": f"Item with ID {item_id} not found"
        }


# DELETE route with path parameter
@app.delete("/api/items/{item_id}")
def delete_item(item_id: int):
    """
    Delete an item by ID.
    Path parameter: item_id (int)
    """
    if item_id in items_db:
        deleted_item = items_db.pop(item_id)
        return {
            "message": f"Item {item_id} deleted successfully",
            "deleted_item": deleted_item
        }
    else:
        return {
            "error": f"Item with ID {item_id} not found"
        }


print("✅ Path parameter routes defined!")
print("Routes created:")
print("  - GET /api/items/{item_id}")
print("  - GET /api/users/{user_id}/posts/{post_id}")
print("  - PUT /api/items/{item_id}")
print("  - DELETE /api/items/{item_id}")
````

## 6️⃣ Section: Add Query Parameters

Query parameters are optional filters passed in the URL after `?`. For example: `/search?q=python&limit=10` where `q` and `limit` are query parameters.

````python
# GET route with optional query parameters
@app.get("/api/products")
def get_products(skip: int = 0, limit: int = 10):
    """
    Get products with pagination.
    Query parameters (optional):
      - skip: Number of products to skip (default: 0)
      - limit: Number of products to return (default: 10)
    Example: GET /api/products?skip=5&limit=20
    """
    all_products = [
        {"id": i, "name": f"Product {i}", "price": i * 10.99}
        for i in range(1, 101)
    ]
    return {
        "skip": skip,
        "limit": limit,
        "products": all_products[skip:skip+limit],
        "total": len(all_products)
    }


# GET route with required query parameter
@app.get("/api/search")
def search_items(q: str, category: str = "all", in_stock: bool = True):
    """
    Search for items.
    Query parameters:
      - q: Search query (REQUIRED)
      - category: Product category (optional, default: "all")
      - in_stock: Filter by stock availability (optional, default: True)
    Example: GET /api/search?q=laptop&category=electronics&in_stock=true
    """
    return {
        "query": q,
        "category": category,
        "in_stock": in_stock,
        "results": [
            {"id": 1, "name": f"Result matching '{q}'", "category": category}
        ],
        "total_results": 1
    }


# GET route combining path and query parameters
@app.get("/api/users/{user_id}/posts")
def get_user_posts(user_id: int, skip: int = 0, limit: int = 5, sort: str = "date"):
    """
    Get posts from a specific user with filtering and sorting.
    Path parameter:
      - user_id: The user ID
    Query parameters:
      - skip: Number of posts to skip (default: 0)
      - limit: Number of posts to return (default: 5)
      - sort: Sort by 'date' or 'likes' (default: 'date')
    Example: GET /api/users/42/posts?skip=10&limit=20&sort=likes
    """
    return {
        "user_id": user_id,
        "skip": skip,
        "limit": limit,
        "sort_by": sort,
        "posts": [
            {"id": i, "title": f"Post {i} from User {user_id}"}
            for i in range(skip, skip+limit)
        ]
    }


print("✅ Query parameter routes defined!")
print("Routes created:")
print("  - GET /api/products?skip=0&limit=10")
print("  - GET /api/search?q=string&category=all&in_stock=true")
print("  - GET /api/users/{user_id}/posts?skip=0&limit=5&sort=date")
````

## 7️⃣ Section: Request Body Validation with Pydantic Models

Pydantic models provide automatic validation and documentation for request bodies. Let's explore advanced Pydantic features.

````python
# Advanced Pydantic model with validation
from pydantic import Field, validator

class Product(BaseModel):
    """Advanced Product model with validation"""
    name: str = Field(..., min_length=1, max_length=100, description="Product name")
    price: float = Field(..., gt=0, description="Price must be greater than 0")
    description: Optional[str] = Field(None, max_length=500, description="Product description")
    quantity: int = Field(default=0, ge=0, description="Quantity must be >= 0")
    is_active: bool = Field(default=True, description="Is product active?")
    tags: List[str] = Field(default=[], description="List of tags")
    
    class Config:
        # Example for documentation
        json_schema_extra = {
            "example": {
                "name": "Gaming Laptop",
                "price": 1299.99,
                "description": "High-performance laptop for gaming",
                "quantity": 5,
                "is_active": True,
                "tags": ["electronics", "computers", "gaming"]
            }
        }


# Another Pydantic model for Order
class OrderItem(BaseModel):
    """Item in an order"""
    product_id: int
    quantity: int = Field(..., gt=0, description="Quantity must be > 0")
    price: float = Field(..., gt=0)


class Order(BaseModel):
    """Order model with nested items"""
    customer_name: str
    email: str
    items: List[OrderItem] = Field(..., min_items=1, description="At least one item required")
    shipping_address: str
    total_price: float = Field(..., ge=0)
    
    class Config:
        json_schema_extra = {
            "example": {
                "customer_name": "John Doe",
                "email": "john@example.com",
                "items": [
                    {"product_id": 1, "quantity": 2, "price": 29.99},
                    {"product_id": 3, "quantity": 1, "price": 79.99}
                ],
                "shipping_address": "123 Main St, New York, NY 10001",
                "total_price": 139.97
            }
        }


# POST route with advanced validation
@app.post("/api/products")
def create_product(product: Product):
    """
    Create a new product with validation.
    All fields are validated according to the Product model.
    """
    return {
        "message": "Product created successfully",
        "product": product,
        "status": "created"
    }


# POST route with nested models
@app.post("/api/orders")
def create_order(order: Order):
    """
    Create a new order with multiple items.
    The order must contain at least one item.
    All prices must be greater than 0.
    """
    return {
        "message": "Order created successfully",
        "order": order,
        "order_id": 12345,
        "status": "pending"
    }


print("✅ Advanced Pydantic models defined!")
print("Models created:")
print("  - Product: with Field validation")
print("  - OrderItem: item in order")
print("  - Order: with nested OrderItem models")
print("\nValidation features used:")
print("  - Field constraints (min_length, max_length, gt, ge, etc.)")
print("  - Optional and List types")
print("  - JSON schema examples")
print("  - Nested models")
````

## 8️⃣ Section: Run the FastAPI Server

To run the FastAPI server, you would normally use Uvicorn from the command line. However, in this notebook, we'll show you the command and then use the `TestClient` for testing instead.

````python
# To run the server in a terminal, use one of these commands:

# Command to run the FastAPI server
commands = """
COMMAND 1: Basic server start (development)
    uvicorn main:app --reload
    
    - main: Your Python file name (main.py)
    - app: The FastAPI application instance variable
    - --reload: Auto-restart on code changes (development only)

COMMAND 2: Specify host and port
    uvicorn main:app --host 0.0.0.0 --port 8000
    
    - --host 0.0.0.0: Listen on all network interfaces
    - --port 8000: Run on port 8000 (default)

COMMAND 3: Production with multiple workers
    uvicorn main:app --workers 4
    
    - --workers 4: Use 4 worker processes (for production)

COMMAND 4: Custom configuration
    uvicorn main:app --reload --log-level debug
    
    - --log-level debug: Show debug logs
"""

print(commands)

# Access the API in your browser or client:
print("\n" + "="*60)
print("ACCESSING YOUR API")
print("="*60)
print("""
Once the server is running at http://localhost:8000:

1. API Endpoints:
   - GET /
   - GET /api/status
   - GET /api/items
   - GET /api/items/{item_id}
   - POST /api/items
   
2. Interactive Documentation:
   - Swagger UI: http://localhost:8000/docs
   - ReDoc: http://localhost:8000/redoc
   - OpenAPI JSON: http://localhost:8000/openapi.json

3. Testing:
   - Browser: http://localhost:8000/
   - Command line: curl http://localhost:8000/api/status
   - Python: requests.get("http://localhost:8000/api/status")
""")
````
